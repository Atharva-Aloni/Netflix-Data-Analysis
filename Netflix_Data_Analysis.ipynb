```python
list_unit = []
list_int = []
```

```python
for i in df.duration:
    if("min" in i):
        list_unit.append("min")
        st = i.replace("min","").strip()
        list_int.append(st)
    elif("Seasons" in i):
        list_unit.append("Seasons")
        st=i.replace("Seasons","").strip()
        list_int.append(st)
    elif("Season" in i):
        list_unit.append("Season")
        st=i.replace("Season","").strip()
        list_int.append(st)
        
```

```python
len(list_int) , len(list_unit)
```

```python
df.duration = list_int
```

```python
df['duration_unit'] = list_unit
```

```python
df
```

```python
df.duration = df.duration.astype('int64')
```

```python
df
```

```python
df_movie = df[df['type']=='Movie']
```

```python
df_movie
```

```python
dm=df_movie.groupby('release_year')['title'].count().reset_index()
```

```python
dm
```

```python
dm_2010_2020 = dm[(dm.release_year>=2010) & (dm.release_year<=2020)]
```

```python
dm_2010_2020
```

```python
sb.barplot(x=dm_2010_2020.release_year,y=dm_2010_2020.title,palette='Set2')
plt.xticks(rotation=90)
plt.ylabel('count of movies ')
plt.xlabel('Release Year')
plt.title('Graph for 2010 to 2020')
plt.show()
```

```python
dm
```

```python
s = dm.title.sum()
year_percent = []
for i in dm.title:
    year_percent.append((i/s)*100)
```

```python
plt.figure(figsize=(20,10))
plt.pie(year_percent,labels=dm.release_year,autopct='%1.1f%%',shadow=True)
plt.show()
```
